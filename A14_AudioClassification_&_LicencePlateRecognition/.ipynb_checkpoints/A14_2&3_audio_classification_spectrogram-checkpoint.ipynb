```python
# helper function for preprocessing

import wave
from pydub import AudioSegment
import soundfile as sf
import noisereduce as nr
import librosa
import numpy as np
import webrtcvad


def is_stereo(wav_file_path):
    with wave.open(wav_file_path, 'rb') as wav_file:
        if wav_file.getnchannels() == 2:
            return True  # It's stereo
        else:
            return False  # It's mono

def count_sterio_mono(dir):
    num_sterios=0
    num_monos=0
    for folder_name in os.listdir(dir):
      folder_path = os.path.join(dir, folder_name)
      for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if is_stereo(file_path):
            # print(f"The file {wav_file_path} is stereo.")
            num_sterios+=1
        else:
            # print(f"The file {wav_file_path} is mono.")
            num_monos+=1
    print(f'num sterios: {num_sterios}')
    print(f'num monos: {num_monos}')

def convert_mono_to_sterio(dir):
    for folder_name in os.listdir(dir):
      folder_path = os.path.join(dir, folder_name)
      for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if not is_stereo(file_path):
          mono_audio = AudioSegment.from_wav(file_path)
          stereo_audio = mono_audio.set_channels(2)
          stereo_audio.export(file_path, format="wav")

def convert_sterio_to_mono(dir):
    for folder_name in os.listdir(dir):
      folder_path = os.path.join(dir, folder_name)
      for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if is_stereo(file_path):
          sterio_audio = AudioSegment.from_wav(file_path)
          mono_audio = sterio_audio.set_channels(1)
          mono_audio.export(file_path, format="wav")

def save_audio_lengths(dir):
    num_frames = []
    for folder_name in os.listdir(dir):
      folder_path = os.path.join(dir, folder_name)
      for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        mono_audio = AudioSegment.from_wav(file_path)
        num_frames.append(   (len(mono_audio.raw_data) // (mono_audio.sample_width * mono_audio.channels)) // mono_audio.frame_rate )
    return num_frames

def denoise(dir):
    for folder_name in os.listdir(dir):
      folder_path = os.path.join(dir, folder_name)
      for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if file_name.endswith(('.wav', '.mp3', '.flac')):  # Check for supported audio formats
            # Load the audio file
            data, rate = librosa.load(file_path)
            # Reduce noise using noisereduce
            reduced_noise = nr.reduce_noise(y=data, sr=rate)
            # Save the denoised audio
            sf.write(file_path, reduced_noise, rate)

      print(f'folder: {folder_name} , denoised')

def delete_lengthy_data(dir, max_duration):
  for folder_name in os.listdir(dir):
      folder_path = os.path.join(dir, folder_name)
      for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if file_name.endswith(('.wav', '.mp3', '.flac')):  # Check for supported audio formats
            # Load the audio file
            data, rate = librosa.load(file_path)
            duration = len(data) / rate
            # Check if duration exceeds max_duration
            if duration > max_duration:
                # Remove the file
                os.remove(file_path)
                # print(f"Removed file: {file_path}")
```

### 3d step: Making spectrograms of the data, and training cnn on it

```python
def get_spectrogram(waveform):
  # Convert the waveform to a spectrogram via a STFT.
  spectrogram = tf.signal.stft(
      waveform, frame_length=255, frame_step=128)
  # Obtain the magnitude of the STFT.
  spectrogram = tf.abs(spectrogram)
  # Add a `channels` dimension, so that the spectrogram can be used
  # as image-like input data with convolution layers (which expect
  # shape (`batch_size`, `height`, `width`, `channels`).
  spectrogram = spectrogram[..., tf.newaxis]
  return spectrogram


def plot_spectrogram(spectrogram, ax):
  if len(spectrogram.shape) > 2:
    assert len(spectrogram.shape) == 3
    spectrogram = np.squeeze(spectrogram, axis=-1)
  # Convert the frequencies to log scale and transpose, so that the time is
  # represented on the x-axis (columns).
  # Add an epsilon to avoid taking a log of zero.
  log_spec = np.log(spectrogram.T + np.finfo(float).eps)
  height = log_spec.shape[0]
  width = log_spec.shape[1]
  X = np.linspace(0, np.size(spectrogram), num=width, dtype=int)
  Y = range(height)
  ax.pcolormesh(X, Y, log_spec)

def make_spec_ds(ds):
  return ds.map(
      map_func=lambda audio,label: (get_spectrogram(audio), label),
      num_parallel_calls=tf.data.AUTOTUNE)
```

```python
train_spectrogram_ds = make_spec_ds(train_ds)
val_spectrogram_ds = make_spec_ds(val_ds)
test_spectrogram_ds = make_spec_ds(test_ds)
```

```python
train_spectrogram_ds.element_spec
```

```python
for example_spectrograms, example_spect_labels in train_spectrogram_ds.take(1):
  break

rows = 3
cols = 3
n = rows*cols
fig, axes = plt.subplots(rows, cols, figsize=(16, 9))

for i in range(n):
    r = i // cols
    c = i % cols
    ax = axes[r][c]
    plot_spectrogram(example_spectrograms[i].numpy(), ax)
    ax.set_title(label_names[example_spect_labels[i].numpy()])

plt.show()
```

```python
# train_ds = train_ds.cache().shuffle(1000).prefetch(tf.data.AUTOTUNE)
# val_ds = val_ds.cache().prefetch(tf.data.AUTOTUNE)
```

```python
input_shape = example_spectrograms.shape[1:]
print('Input shape:', input_shape)
num_labels = len(label_names)

# Instantiate the `tf.keras.layers.Normalization` layer.
norm_layer = layers.Normalization()
# Fit the state of the layer to the spectrograms
# with `Normalization.adapt`.
norm_layer.adapt(data=train_spectrogram_ds.map(map_func=lambda spec, label: spec))
```

```python
num_labels=10

model = tf.keras.models.Sequential([
    layers.Input(shape=(None, None, 1)),
    # Downsample the input.
    layers.Resizing(32, 32, 'nearest'),
    # Normalize.
    norm_layer,
    layers.Conv2D(32, 3, activation='relu'),
    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(128, 3, activation='relu'),
    # layers.MaxPooling2D(),
    layers.Dropout(0.2),
    # layers.Flatten(),
    layers.GlobalAveragePooling2D(),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(num_labels),
])

model.summary()

model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy'],
)
```

```python
 early_stopping= tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                                            patience=20,
                                            restore_best_weights=True)

reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=4)

EPOCHS = 100

history = model.fit(
    train_spectrogram_ds,
    validation_data=val_spectrogram_ds,
    epochs=EPOCHS,
    callbacks=[early_stopping, reduce_lr],
)
```

### 4th step: Python script to get audio path and tell the class

```python
# python script to get audio path and tell the class

# I don't write it in a seperate .py file, 
# as the model does not load properly in my system
# I just run it here on colab

import tensorflow as tf
import numpy as np
import librosa
import soundfile as sf

# Load the saved model
model_export = tf.keras.models.load_model('drive/MyDrive/audio_classification_spectrogram.keras')

# Function to convert audio to spectrogram
def get_spectrogram(waveform):
    spectrogram = tf.signal.stft(
        waveform, frame_length=255, frame_step=128)
    spectrogram = tf.abs(spectrogram)
    spectrogram = spectrogram[..., tf.newaxis]
    return spectrogram

# Function to convert audio file to spectrogram
def audio_to_spectrogram(file_path):
    # Load audio file
    audio, sr = librosa.load(file_path)
    # Ensure audio is mono
    if len(audio.shape) > 1:
        audio = np.mean(audio, axis=1)

    # Pad or trim audio to 2 seconds
    if len(audio) < 2 * 22050:
        audio = np.pad(audio, (0, 2 * 22050 - len(audio)), mode='constant')
    elif len(audio) > 2 * 22050:
        audio = audio[:2 * 22050]

    # Convert audio to spectrogram
    spectrogram = get_spectrogram(audio)

    # Resize spectrogram to match model input size
    resized_spectrogram = tf.image.resize(spectrogram, (32, 32))

    return resized_spectrogram

# Function to classify audio file
def classify_audio(file_path):
    spectrogram = audio_to_spectrogram(file_path)
    spectrogram = tf.expand_dims(spectrogram, axis=0)  # Add batch dimension

    # Normalize spectrogram (if needed)
    # spectrogram = spectrogram / tf.reduce_max(spectrogram)

    prediction = model_export.predict(spectrogram)
    class_index = np.argmax(prediction)
    return class_index

# # Get file path from user
# file_path = input("Enter the path to your audio file: ")

# # Classify the audio file
# class_index = classify_audio(file_path)

# # Print the class of the audio file
# print(f"The class of the audio file is: {class_index}")
```

### Python script to get audio from microphone and tell the class

One solution to record a file is sounddevice

Another solution to record a file is PvRecorder

```python
import tensorflow as tf
import numpy as np
from pvrecorder import PvRecorder
import librosa

model_export = tf.keras.models.load_model('drive/MyDrive/audio_classification_spectrogram.keras')

# function to convert audio to spectrogram
def get_spectrogram(waveform):
    spectrogram = tf.signal.stft(
        waveform, frame_length=255, frame_step=128)
    spectrogram = tf.abs(spectrogram)
    spectrogram = spectrogram[..., tf.newaxis]
    return spectrogram

# function to record audio from microphone
def record_audio_from_microphone(seconds=2):
    # Parameters
    frame_length = 512  # 32 ms at 16 kHz

    # Create PvRecorder instance
    recorder = PvRecorder(device_index=-1, frame_length=frame_length)

    print("Recording...")
    frames = []
    recorder.start()
    for i in range(int(seconds * 1000 / frame_length)):
        frame = recorder.read()
        frames.append(frame)

    print("Finished recording")

    # stop and delete recorder
    recorder.stop()
    recorder.delete()

    # convert frames to a single numpy array
    audio = np.concatenate(frames, axis=0)

    # sample audio to 22,050 Hz
    audio = librosa.resample(audio, orig_sr=16000, target_sr=22050)

    return audio

# function to process recorded audio
def process_recorded_audio(audio):
    # ensure audio is mono (already done by PvRecorder)
    # pad or trim audio to 2 seconds
    if len(audio) < 2 * 22050:
        audio = np.pad(audio, (0, 2 * 22050 - len(audio)), mode='constant')
    elif len(audio) > 2 * 22050:
        audio = audio[:2 * 22050]

    spectrogram = get_spectrogram(audio)

    resized_spectrogram = tf.image.resize(spectrogram, (32, 32))

    return resized_spectrogram

def classify_recorded_audio(audio):
    spectrogram = process_recorded_audio(audio)
    spectrogram = tf.expand_dims(spectrogram, axis=0)  # Add batch dimension

    prediction = model_export.predict(spectrogram)
    class_index = np.argmax(prediction)
    return class_index

audio = record_audio_from_microphone()

class_index = classify_recorded_audio(audio)

print(f"The class of the recorded audio is: {class_index}")
```
